# Midwestern US states earthquake frequency over the last 30 years

The data was obtained from USGS here: http://earthquake.usgs.gov/earthquakes/search/. It was then run through a PostGIS database to determine the location of the epicenters by state, since USGS does provide specific state locations in its data.

```python
import matplotlib.pyplot as plt
import pandas as pd

all_quakes = pd.DataFrame.from_csv("earthquakes.csv", index_col=None, parse_dates=["time", "updated"])
len(all_quakes)
```

### Years spanned?

```python
min(all_quakes.time)
```

```python
max(all_quakes.time)
```

### Which states?

```python
list(set(all_quakes['state']))
```

## Oklahoma earthquakes

```python
ok_quakes = all_quakes.select("state", "Oklahoma")
len(ok_quakes)
```

Resample the time-series data and sum by year, then plot.

```python
data = ok_quakes.set_index("time")["mag"].resample("A", how="count")
plt.plot(data)
```

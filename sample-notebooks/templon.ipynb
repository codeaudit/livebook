## Analyzing State-by-State Changes In Earthquake Frequency

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Load and clean up data

The data was obtained from USGS here: http://earthquake.usgs.gov/earthquakes/search/. It was then run through a PostGIS database to determine the location of the epicenters by state, since USGS does provide specific state locations in its data.

```python
all_quakes = pd.DataFrame.from_csv("earthquakes.csv", index_col=None, parse_dates=["time", "updated"])
len(all_quakes)
```

## Charting Earthquake Activity Over Time, By State

```python
ok_quakes = all_quakes.select("state", "Oklahoma")
data = ok_quakes.set_index("time")["id"].resample("A", how="count")
plt.plot(data)
```

```python
ok_quakes = all_quakes.select("state", "Maine")
data = ok_quakes.set_index("time")["id"].resample("A", how="count")
plt.plot(data)
```

```python
ok_quakes = all_quakes.select("state", "Arizona")
data = ok_quakes.set_index("time")["id"].resample("A", how="count")
plt.plot(data)
```
